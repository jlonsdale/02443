```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom
from scipy.stats import norm
from scipy.stats import pareto
import random
import math
```

```python
def exp_dist(l):
    return -np.log(random.uniform(0,1))/l
```

```python
results =[]
l=0.5
for i in range(0,1000):
    results.append(exp_dist(l))

plt.hist(results, align='left', rwidth=0.8)
plt.xlabel('Number of Trials')
plt.ylabel('Frequency')
plt.show()
```

```python
z1=[]
z2=[]



def box_muller():
    random_value1 = random.uniform(0,1)
    random_value2 = random.uniform(0,1)
    theta=2*np.pi*random_value2
    r = math.sqrt(-2*np.log(random_value1))
    z1.append(r*math.cos(theta))
    z2.append(r*math.cos(theta))

for i in range(0,1000):
    box_muller() 

plt.hist(z1, align='left', rwidth=0.8, density=True)
plt.hist(z2, align='left', rwidth=0.7, density=True)
x = np.linspace(norm.ppf(0.01),
                norm.ppf(0.99), 1000)
plt.plot(x, norm.pdf(x),
       'r-', lw=1, alpha=0.6, label='norm pdf')
plt.xlabel('Number of Trials')
plt.ylabel('Frequency')
plt.show()
```
